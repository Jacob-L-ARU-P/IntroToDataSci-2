# Pre-processing the data
## Removing comma separators
This data has already been pre-processed from the original data set. The following has been done:
* Removing all the comma separators in the numerical fields so they import as integers or floats.
* Combining the data sets for 2001 and 2011 and adding a feature for `Year`.

## Converting `Year` to a category
In this activity you will explore the differences between 2001 and 2011. It will be helpful if `Year` is reconfigured as a categorical feature to use many of the built-in commands in Seaborn. To do this you can set the data type as *category* using: `travel_data['Year'].astype('category')`

* Run the code below to change the data type of `Year` to a category.

```python
# change the data type of Year to string
travel_data['Year'] = travel_data['Year'].astype('category')

# display the data types and info
travel_data.info()
```
